```r
Data_Frame<-data.frame(
    id=c(1,2,3,4,5),
    name=c('Peter', 'Amy', 'Ryan', 'Gary', 'Michelle'),
    salary=c(623.30,515.20,611.00,729.00,843.25)
)
Data_Frame
new_col_DF<-cbind(Data_Frame, Department=c('IT', 'HR', 'Management', 'Audit','Internal'))
new_col_DF
```

```r
Data_Frame_minus1<-new_col_DF[-c(1,3,5), -c(2,3)]
Data_Frame_minus1
```

```r
Data_Frame_minus2<-new_col_DF[-c(2,3), ]
Data_Frame_minus2
```

```r
barplot(Data_Frame_minus2$salary, names.arg=Data_Frame_minus2$name)
```

```r
max<-max(new_col_DF$salary)
min<-min(new_col_DF$salary)
median<-median(new_col_DF$salary)
x<-c(max,min,median)
mylabel<-c('Highest','Lowest','Median')
pie(x, label=mylabel, main="Salary")
```

```r
#monthly summary
df<-read.csv('deenaamazon.csv')
library(lubridate)
library(dplyr)
Date = as.Date(df$Order.Date, format='%m/%d/%Y')
df%>%
    group_by(month=floor_date("Order.Date", "month")) %>%
    summarize(mean(df$"Total.Charged"))
```

```r
df<-read.csv('deenaamazon.csv')
df[is.na(df)]<-0

df$"Tax.Charged"=as.numeric(gsub("\\$", "", df$"Tax.Charged"))
barplot(df$"Tax.Charged", names.arg=df$"Order.Date", main="Tax Overtime")

df$"Subtotal"=as.numeric(gsub("\\$", "", df$"Subtotal"))
barplot(df$"Subtotal", names.arg=df$"Order.Date", main="Subtotal Overtime")
```

```r
df<-read.csv('deenaamazon.csv')
df[is.na(df)]<-0
df$"Subtotal"=as.numeric(gsub("\\$", "", df$"Subtotal"))
df$"Tax.Charged"=as.numeric(gsub("\\$", "", df$"Tax.Charged"))

total<-sum(df$"Subtotal")
tax<-sum(df$"Tax.Charged")

x<-c(total, tax)
mylabel<-c("total", "tax")
pie(x, label=mylabel, main="Tax to Subtotal Spending Proportion")
```
